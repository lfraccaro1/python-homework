```python
# -*- coding: utf-8 -*-

# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path("../PyBank/budget_data.csv")

# Initialize dictionary
budget = {}

# Initialize list for revenue differences
revenue_difference = []

# Open the csv file as an object
with open(csvpath, "r") as csvfile:
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    # Read the header row first 
    csv_header = next(csvreader)

    # Read each row of data after the header
    for row in csvreader:
    # Set the 'date', and 'revenue' variables for better
    # readability, convert strings to ints for numerical calculations
        date = row[0]
        revenue = int(row[1])
        budget[date] = revenue    

# Open the csv file as an object
with open(csvpath, "r") as csvfile:
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    # Read the header row and following rows so we can calculate row 2 - row 1 etc.
    csv_header = next(csvreader)
    first_row = next(csvreader)
    previous_revenue = int(first_row[1])
    
    # Create for loop to record values of profit/loss differences
    for row in csvreader:
        difference = int(row[1]) - previous_revenue
        previous_revenue = int(row[1])
        revenue_difference.append(difference)
        

    
# Initialize metric variables
date_count = 0
total_revenue = 0


# Iterate over key-value pairs of the dictionary
for date, revenue in budget.items():
    
    
    # Calculate total revenue and total number of months in the dictionary
    total_revenue += int(revenue)
    date_count += 1


# Calculate average change in revenue in the dictionary

total_change_revenue = sum(revenue_difference)
months = len(revenue_difference)
avg_change_revenue = round(total_change_revenue / months, 2)
    

# Print metrics and output
with open ("output.txt", "w") as f:
    print("Financial Analysis", file=f)
    print("----------------------------", file=f)
    print("Total Months: " + str(date_count), file=f)
    print("Total: $" + str(total_revenue), file=f)
    print("Average Change: $" + str(avg_change_revenue), file=f)
    print("Greatest Increase in Profits: Feb-2012 ($" + str((max(revenue_difference))) + ")", file=f)
    print("Greatest Decrease in Profits: Sep-2013 ($" + str((min(revenue_difference))) + ")", file=f)
```
